```python
from datetime import datetime

def convert_months(data, col):
  ref_date = datetime(2016, 1, 1)
  time_delta = ref_date - data[col]
  # Calculate months by converting timedelta to days and dividing by average days in a month
  # Return NaN for NaT values
  m = (time_delta / np.timedelta64(1, 'D') / 30.44)
  return m
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import pandas as pd

data = data.drop_duplicates().dropna()

train_data = data.copy()

columns_to_drop = ['id', 'churn', 'date_activ', 'date_end', 'date_modif_prod', 'date_renewal', 'tenure']
X = data.drop(columns = columns_to_drop, axis = 1)
y = data['churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 42)

nan_in_y_train = y_train.isna()
X_train = X_train[~nan_in_y_train]
y_train = y_train[~nan_in_y_train]

numerical_features = X_train.select_dtypes(include=np.number).columns
categorical_features = X_train.select_dtypes(include='bool').columns

numerical_transformer = StandardScaler()
categorical_transformer = 'passthrough'

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='passthrough'
)

X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)


model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_processed, y_train)

y_pred = model.predict(X_test_processed)
```

```python
accuracy_score(y_test, y_pred)
```

```python
class_report = classification_report(y_test, y_pred)
print(class_report)
```

```python
fig, axes = plt.subplots(1, 2, figsize = (18, 8), constrained_layout = True) # Adjusted figure size for better readability

feat_imp = pd.DataFrame({'feature': X_train.columns, 'importance': model.feature_importances_})
feat_imp = feat_imp.sort_values(by = 'importance', ascending = True)

# Select the top 15 features
top_15_features = feat_imp.tail(40)

sns.histplot(y_test - y_pred, kde = True, ax = axes[0])

# Plot the top 15 features
axes[1].barh(range(len(top_15_features)), top_15_features['importance'], tick_label = top_15_features['feature'])
axes[1].set_title('Top 40 Feature Importances') # Add a title for clarity
axes[1].set_xlabel('Importance') # Add x-label
```

```python
proba_predictions = model.predict_proba(X_test)
probabilities = proba_predictions[:, 1]
probabilities
```
